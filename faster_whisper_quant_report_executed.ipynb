# Faster-Whisper Quantization Experiments

Этот notebook содержит только эксперименты по квантизации для `faster-whisper`.

Успешно отработали quant-only варианты:
- `faster-whisper|CPU|int8`
- `faster-whisper|CPU|int8_float32`

Дополнительно были попробованы GPU-варианты:
- `faster-whisper|GPU|int8_float16`
- `faster-whisper|GPU|int8`

Обе GPU-конфигурации на этом сервере завершились `CUDA out of memory`, поэтому итоговая таблица `RTF` ниже содержит только успешные quant-only запуски.

Все ячейки выполнены, а тяжелый прогон был сделан заранее на сервере.
